```python
import os

print("=== INVESTIGA√á√ÉO COMPLETA ===")

# Verifica o diret√≥rio atual do Jupyter
diretorio_atual = os.getcwd()
print(f"üìÇ Diret√≥rio atual do Jupyter: {diretorio_atual}")

# Lista ABSOLUTAMENTE TUDO neste diret√≥rio
print(f"\nüìÑ Conte√∫do de {diretorio_atual}:")
try:
    arquivos = os.listdir(diretorio_atual)
    for arquivo in arquivos:
        print(f"   {arquivo}")
except Exception as e:
    print(f"‚ùå Erro ao listar arquivos: {e}")

# Procura especificamente pelo arquivo .env em v√°rios locais
print(f"\nüîç Procurando por .env_kafka_connect...")

locais_possiveis = [
    '/home/jovyan/.env_kafka_connect',  # Local mais prov√°vel
    '/home/jovyan/env_kafka_connect',
    '.env_kafka_connect',
    '../.env_kafka_connect',
    '/app/.env_kafka_connect',
    '/opt/.env_kafka_connect',
    diretorio_atual + '/.env_kafka_connect'
]

encontrado = False
for caminho in locais_possiveis:
    if os.path.exists(caminho):
        print(f"‚úÖ ENCONTRADO em: {caminho}")
        print(f"   Tamanho: {os.path.getsize(caminho)} bytes")
        
        # Mostra o conte√∫do
        with open(caminho, 'r') as f:
            conteudo = f.read()
        print(f"   Conte√∫do:\n{conteudo}")
        
        encontrado = True
        break

if not encontrado:
    print("‚ùå N√ÉO ENCONTRADO em nenhum local")
    
    # Mostra o conte√∫do de /home/jovyan para ver o que tem l√°
    print(f"\nüëÄ Conte√∫do de /home/jovyan:")
    try:
        for item in os.listdir('/home/jovyan'):
            print(f"   {item}")
    except Exception as e:
        print(f"   N√£o foi poss√≠vel acessar /home/jovyan: {e}")
```

```python
import os

print("=== PROCURANDO JARS ===")

# Locais comuns onde os JARs podem estar
locais_jars = [
    '/opt/spark/jars',           # Pasta padr√£o do Spark
    '/home/jovyan/jars',         # Pasta no home
    '/jars',                     # Pasta raiz
    '/app/jars',                 # Pasta app
    '../jars',                   # Uma pasta acima
    './jars'                     # Pasta atual
]

for local in locais_jars:
    if os.path.exists(local):
        print(f"‚úÖ Pasta encontrada: {local}")
        arquivos = os.listdir(local)
        jars = [f for f in arquivos if f.endswith('.jar')]
        print(f"   üì¶ JARs encontrados: {len(jars)}")
        for jar in jars[:5]:  # Mostra os 5 primeiros
            print(f"      - {jar}")
        if len(jars) > 5:
            print(f"      ... e mais {len(jars) - 5} JARs")
        print()
    else:
        print(f"‚ùå Pasta n√£o existe: {local}")
```
